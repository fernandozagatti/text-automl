# Using limit of tokens
* Terms will have up to the assigned number, use the hyperparameter "number_of_tokens".

```python
import sys
sys.path.append('../terms_extraction/')

import pandas as pd
from terms_extractor import terms_extractor

df = pd.read_csv('../datasets/B2W-Reviews01.csv')
df = df[['product_name']]

extract, processed_df = terms_extractor(data=df, measures=['yake', 'tfidf', 'pmi', 'mle', 'dice', 'count'], 
                                        target_column='product_name', number_of_tokens=3, verbose=1)

display(processed_df)
```

# Specifying number of terms to extract
* Use the hyperparameter "number_to_extract".

```python
import sys
sys.path.append('../terms_extraction/')

import pandas as pd
from terms_extractor import terms_extractor

df = pd.read_csv('../datasets/B2W-Reviews01.csv')
df = df[['product_name']]

extract, processed_df = terms_extractor(data=df, measures=['yake', 'tfidf', 'pmi', 'mle', 'dice', 'count'], 
                                        target_column='product_name', number_of_tokens=3, 
                                        number_to_extract=500, verbose=1)

display(processed_df)
```

# Specifying number of terms to extract by threshould
* The hyperparameter "threshold_percentage" will define the number of terms to extract based in the number of n-grams.

```python
import sys
sys.path.append('../terms_extraction/')

import pandas as pd
from terms_extractor import terms_extractor

df = pd.read_csv('../datasets/B2W-Reviews01.csv')
df = df[['product_name']]

extract, processed_df = terms_extractor(data=df, measures=['yake', 'tfidf', 'pmi', 'mle', 'dice', 'count'], 
                                        target_column='product_name', number_of_tokens=3,
                                        threshold_percentage=0.995, verbose=1)

display(processed_df)
```

# Using your own list

```python
import sys
sys.path.append('../terms_extraction/')

import pandas as pd
from terms_extractor import terms_extractor

df = pd.read_csv('../datasets/B2W-Reviews01.csv')
df = df[['product_name']]

extract=['fita led', 'painel', 'mop giratorio', 'carregador', 'notebook', 'smart tv', 'copo']

extract, processed_df = terms_extractor(data=df, extract=extract, target_column='product_name', 
                                        number_of_tokens=2, stopwords=True, language='portuguese', 
                                        verbose=1)

display(processed_df)
```

# Showing other functionalities...
* For example, it is possible to use a list of names to identify texts that need to be anonymized.

```python
import sys
sys.path.append('../terms_extraction/')

import pandas as pd
from terms_extractor import terms_extractor

df = pd.DataFrame(
    {'review_text': ['Meu nome é Pedro e odiei o serviço', 
                     'A atendente Márcia foi muito boa!', 
                     'Adorei a entrega']
    }
)

_, processed_df = terms_extractor(data=df, target_column='review_text', 
                                  extract=['pedro', 'marcia'], number_of_tokens=1)

display(processed_df)
```
